### 6. Implement a softmax (multinomial) logistic regression model from scratch. [Marks 1]

```python
# Solution Code
class SoftmaxRegression:
    def __init__(self, learning_rate=0.01, iterations=1000, reg_param=0.01):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.reg_param = reg_param
        self.weights = None
        self.bias = None
        self.loss_history_train = []
        self.loss_history_val = []

    def softmax(self, z):
        exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
        return exp_z / np.sum(exp_z, axis=1, keepdims=True)

    def cross_entropy_loss(self, y_true, y_pred):
        epsilon = 1e-15
        y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
        loss = -np.mean(np.sum(y_true * np.log(y_pred), axis=1))
        loss += (self.reg_param / 2) * np.sum(self.weights**2)
        return loss

    def fit(self, X_train, y_train, X_val, y_val):
        m, n = X_train.shape
        k = y_train.shape[1]
        self.weights = np.zeros((n, k))
        self.bias = np.zeros(k)

        for i in range(self.iterations):
            z_train = X_train.dot(self.weights) + self.bias
            y_pred_train = self.softmax(z_train)
            loss_train = self.cross_entropy_loss(y_train, y_pred_train)
            self.loss_history_train.append(loss_train)
            z_val = X_val.dot(self.weights) + self.bias
            y_pred_val = self.softmax(z_val)
            loss_val = self.cross_entropy_loss(y_val, y_pred_val)
            self.loss_history_val.append(loss_val)
            gradient_weights = (1 / m) * X_train.T.dot(y_pred_train - y_train) + self.reg_param * self.weights
            gradient_bias = (1 / m) * np.sum(y_pred_train - y_train, axis=0)
            self.weights -= self.learning_rate * gradient_weights
            self.bias -= self.learning_rate * gradient_bias

    def predict_proba(self, X):
        z = X.dot(self.weights) + self.bias
        return self.softmax(z)

    def predict(self, X):
        probabilities = self.predict_proba(X)
        return np.argmax(probabilities, axis=1)

    def get_loss_history(self):
        return self.loss_history_train, self.loss_history_val

# Create and fit the model
model = SoftmaxRegression(learning_rate=0.01, iterations=1000)
model.fit(train_x, train_y, val_x, val_y)
```

### 7. Using cross entropy as loss and apply L2 regularization to the weights in the cost function and train the softmax logistic regression model with gradient descent. Use cross validation to find a suitable value of the learning rate and regularization parameter. Also give the training and validation loss vs. training epochs plot. [Marks 6]

```python
#solution code
model.cross_entropy_loss(train_y, model.predict_proba(train_x))
```

```python
from sklearn.model_selection import KFold

# Cross-validation to find optimal hyperparameters
learning_rates = [0.001, 0.01, 0.1]
reg_params = [0.001, 0.01, 0.1]
best_loss = float('inf')
best_lr = None
best_reg = None

# Combine training and validation data for cross-validation
X_train_val = np.vstack((train_x, val_x))
y_train_val = np.vstack((train_y, val_y))

kf = KFold(n_splits=5, shuffle=True, random_state=42)

for lr in learning_rates:
    for reg in reg_params:
        fold_losses = []
        for train_index, val_index in kf.split(X_train_val):
            X_fold_train, X_fold_val = X_train_val[train_index], X_train_val[val_index]
            y_fold_train, y_fold_val = y_train_val[train_index], y_train_val[val_index]

            fold_model = SoftmaxRegression(learning_rate=lr, iterations=500, reg_param=reg) # Reduced iterations for faster CV
            fold_model.fit(X_fold_train, y_fold_train, X_fold_val, y_fold_val)
            fold_losses.append(fold_model.loss_history_val[-1])

        avg_loss = np.mean(fold_losses)
        print(f"LR: {lr}, Reg: {reg}, Avg Validation Loss: {avg_loss:.4f}")

        if avg_loss < best_loss:
            best_loss = avg_loss
            best_lr = lr
            best_reg = reg

print(f"\nBest Learning Rate: {best_lr}")
print(f"Best Regularization Parameter: {best_reg}")

# Train the final model with the best hyperparameters
final_model = SoftmaxRegression(learning_rate=best_lr, iterations=1000, reg_param=best_reg)
final_model.fit(train_x, train_y, val_x, val_y)
```

```python
# Plot training and validation loss for the final model
plt.figure(figsize=(6, 4))
plt.plot(final_model.get_loss_history()[0], label='Training Loss', linewidth=2)
plt.plot(final_model.get_loss_history()[1], label='Validation Loss', linestyle='--', linewidth=2)
plt.xlabel('Iterations', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.title('Training and Validation Loss (Final Model)', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=11)
plt.tight_layout()
plt.grid(True)
plt.show()
```

### 8. Report the prediction performance of your implementation on the test set in terms of confusion matrix, precision, recall and area under precision recall curve. [Marks 2]

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, average_precision_score, PrecisionRecallDisplay
import matplotlib.pyplot as plt
import numpy as np

# Convert predictions and true labels
y_pred_proba = final_model.predict_proba(test_x)
y_pred = np.argmax(y_pred_proba, axis=1)
y_test_single = np.argmax(test_y, axis=1)

# Confusion Matrix
cm = confusion_matrix(y_test_single, y_pred)
print("Confusion Matrix:")
print(cm)

# Overall Precision & Recall
precision = precision_score(y_test_single, y_pred, average='macro')
recall = recall_score(y_test_single, y_pred, average='macro')

print(f"\nPrecision: {precision:.4f}")
print(f"Recall: {recall:.4f}")

# Overall AUPRC
average_precision = average_precision_score(test_y, y_pred_proba, average='macro')
print(f"\nArea Under Precision-Recall Curve (AUPRC): {average_precision:.4f}")

# Plot combined Precision-Recall Curve (macro-average)
PrecisionRecallDisplay.from_predictions(
    test_y.ravel(), y_pred_proba.ravel(), name="Macro-Average"
)
plt.title("Precision-Recall Curve (Macro-Average)")
plt.show()
```
